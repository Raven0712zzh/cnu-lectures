```python
from math import exp,expm1

def f0(x):
    return exp(x) - 1

def f1(x):
    return expm1(x)

print(f0(1e-14))    #alot rounding error 
                    # - we got 9.992007221626409e-15 means [scientific notation]9.992...*10^(-15)
print(f1(1e-14))
```

```python
def f(x):
    eps = 1e-5
    if abs(x) < eps:
        return 1/6 + (1/24) *x + (1/120) *x ** 2    # factorial 4！，5！，分子抵消
    else:
        return(exp(x) - 1 - x - 0.5 * x * x) / (x ** 3)
    
print(f(1e-8))
```

$s = \sum_{i=1}^{10} \sqrt{i}$


```python
def s(N):
    v = 0.0
    for i in range(1, N+1):
        v += i ** 0.5
    return v

print (s(10))
```
